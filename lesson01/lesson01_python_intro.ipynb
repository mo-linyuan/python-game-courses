#### 第一件事：让电脑说一句话

**先想一想**

“如果你现在要命令电脑说一句欢迎词，你觉得应该怎么做？”

**答案**

Python 里最常用的显示命令之一，就是 `print()`。

```python
print("欢迎来到 Python 世界！")
```

#### 看懂这行代码

**可以慢慢拆开讲**

- `print` 的意思可以理解成“显示出来”
- `()` 里面放要显示的内容
- 双引号里的文字，会原样显示到屏幕上

**一句最重要的话**

`print()` 的作用就是：把内容显示出来。

#### 小练习：连续说三句话

**目标**

**先记住这一点**

每一行都要写一个完整的 `print()`。

```python
print("大家好")
print("我是今天的主角")
print("我要开始学习 Python 了")
```

#### 为什么代码像在发指令

**可以这样理解**

写程序有点像给一个特别听话、但一点都不会猜的机器人下命令。你写什么，它就做什么；你少写一个符号，它就可能不明白。

#### 程序会按顺序执行

**这部分是本课核心之一**

电脑不是一下子看懂全部代码，而是从上到下，一行一行执行。

**先观察下面这段**

把它当成故事镜头来看。

```python
print("清晨，闹钟响了。")
print("主角一下子坐了起来。")
print("他想起今天要去学习 Python。")
```
