#### 先理解列表是什么

**一句话理解**

列表就像一个收纳盒，里面可以放很多个东西。

如果变量像一个小盒子，那么列表就像一个有很多格子的盒子。

```python
bag = ["木剑", "药水", "火把"]

print(bag)
```

#### 看懂这个列表

**先认识外形**

- 列表用方括号 `[]`
- 里面可以放多个内容
- 每个内容之间用逗号分开

#### 列表里的东西可以单独拿出来

**这一步很重要**

列表里的每个位置都有编号。

Python 里，第一个位置的编号是 `0`，不是 `1`。

```python
bag = ["木剑", "药水", "火把"]

print("第一个物品：", bag[0])
print("第二个物品：", bag[1])
print("第三个物品：", bag[2])
```

#### 列表里的内容可以增加

**今天再学一个很常用的动作**

如果角色找到新道具，就可以把它加进背包。

```python
bag = ["木剑", "药水", "火把"]
bag.append("钥匙")

print(bag)
```

#### 还能看看一共有多少东西

**这里顺便认识一个新工具**

`len()` 可以用来查看列表里有几个元素。

```python
bag = ["木剑", "药水", "火把", "钥匙"]

print("背包里的物品数量：", len(bag))
```
