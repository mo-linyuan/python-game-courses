## 第9课：循环让程序不怕重复

#### 本节课的目的

前一节课里，我们已经会用列表保存很多个物品。

但如果背包里有 5 个物品、10 个物品，一个一个手写 `print()` 就会很麻烦。

这一节课，我们要认识一个非常重要的能力：循环。

循环可以让程序重复做同一件事，这也是电脑最擅长的本领之一。

**这节课结束时，你应该能做到**

- 知道 `for` 循环是做什么的
- 会写最基础的 `for` 循环
- 会让程序重复输出内容
- 能把列表和循环放在一起使用

#### 先看一个没有循环的例子

**先体会一下为什么需要循环**

```python
print("勇者继续前进。")
print("勇者继续前进。")
print("勇者继续前进。")
```

如果要重复 3 次、10 次、100 次，这样写就太麻烦了。

所以我们要让程序自动重复。

```python
for i in range(3):
    print("勇者继续前进。")
```

#### 看懂这段代码

**先抓住重点**

- `for` 表示开始循环
- `range(3)` 表示重复 3 次
- 缩进的那一行会被重复执行

#### 再看一个带回合感的例子

**这时候就更像游戏了**

```python
for i in range(1, 4):
    print("第", i, "回合开始。")
```

#### 列表和循环特别适合一起用

**这就是和上一节课的衔接点**

如果背包里有很多物品，我们可以让程序一个一个显示出来。

```python
bag = ["木剑", "药水", "火把", "钥匙"]

for item in bag:
    print(item)
```

#### 再让它更像背包界面

**循环里不一定只能打印一个变量**

```python
bag = ["木剑", "药水", "火把", "钥匙"]

for item in bag:
    print("背包物品：", item)
```
