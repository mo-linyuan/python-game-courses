#### 先认识字符串

**一句话理解**

字符串就是文字。

只要你想让程序记住一段文字，一般就要把它放在引号里。

```python
title = "森林试炼"
hero = "林风"
weapon = "木剑"

print(title)
print(hero)
print(weapon)
```

#### 看懂这段代码

**可以边讲边指着看**

- `"森林试炼"` 是一段文字
- `title` 是保存这段文字的变量名
- 文字外面的引号不能少

#### 字符串可以拼起来

**这节课的核心本领之一**

游戏里的很多句子，不一定是一开始就完整写死的。

我们可以把几段文字拼起来，组成一句完整的话。

```python
hero = "林风"
weapon = "木剑"

print(hero + " 拿着 " + weapon + " 出发了。")
```

#### 再看一个更像游戏的例子

**这时候学生会更容易代入**

```python
place = "迷雾森林"
task = "找到失落的宝箱"

print("你来到了 " + place + "。")
print("当前任务：" + task)
```

#### 把输入和字符串连起来

**这一步很实用**

如果角色名字由玩家输入，故事就会更像“属于自己”的故事。

```python
name = input("请输入角色名字：")
weapon = input("请输入你的武器：")

print(name + " 拿起了 " + weapon + "。")
print("新的冒险开始了。")
```

#### 一个小提醒

**今天先记住这个就够了**

如果你要拼接的是文字，通常用 `+`。

如果只是想把文字和数字一起显示，暂时最稳妥的写法还是像前面那样在 `print()` 里用逗号分开。

```python
name = "林风"
hp = 80

print(name, "现在的生命值是", hp)
```
