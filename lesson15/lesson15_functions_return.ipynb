## 第15课：让函数把答案交回来——返回值

#### 本节课的目的

前两节课我们已经学会了：

- 第13课：函数可以把重复代码打包
- 第14课：函数可以接收参数

但目前我们写的函数只会做事，不会给结果。

比如你想让函数帮你算伤害，然后把结果拿回来用——光靠 `print` 是做不到的。

这节课，我们学最后一个关键概念：**返回值**。

**这节课结束时，你应该能做到**

- 知道返回值是什么
- 会用 `return` 让函数输出结果
- 会把返回值保存到变量里继续使用
- 能用函数完成战斗伤害计算、存活判断等逻辑

#### 先看一个问题

**这个函数有什么不方便的地方？**

```python
def calculate_damage(attack, defense):
    damage = attack - defense
    print('伤害：', damage)

calculate_damage(20, 5)
```

这个函数只能打印出伤害，但没办法把伤害值交给别的地方用。

比如你想用伤害值去扣生命值，就没法做到。

#### 用 return 把结果交回来

**一句话理解**

`return` 就是函数完成任务后，把答案递出来。

```python
def calculate_damage(attack, defense):
    damage = attack - defense
    return damage

result = calculate_damage(20, 5)
print('造成伤害：', result)
```

#### 看懂这段代码

**拆开来看**

- `return damage` 的意思是：函数执行完后，把 `damage` 的值交出去
- `result = calculate_damage(20, 5)` 把函数交出来的值存到 `result` 里
- 之后可以继续用 `result` 做任何事情

#### 返回值可以直接用

**不一定非要存进变量**

```python
def calculate_damage(attack, defense):
    return attack - defense

# 直接用在 print 里
print('伤害：', calculate_damage(20, 5))

# 或者存起来再用
dmg = calculate_damage(30, 8)
print('暴击伤害：', dmg)
```

#### 真实游戏场景：把返回值用来扣血

**这才是返回值最大的用处**

```python
def calculate_damage(attack, defense):
    damage = attack - defense
    if damage < 0:
        damage = 0
    return damage

player = {'名字': '林风',   '生命值': 100, '攻击力': 18}
enemy  = {'名字': '暗影狼', '生命值': 60,  '攻击力': 12, '防御力': 5}

dmg = calculate_damage(player['攻击力'], enemy['防御力'])
enemy['生命值'] = enemy['生命值'] - dmg

print(player['名字'], '攻击了', enemy['名字'], '，造成', dmg, '点伤害！')
print(enemy['名字'], '剩余生命值：', enemy['生命值'])
```

#### 返回值可以是判断结果

**函数也可以回答「是」或「否」**

```python
def is_alive(character):
    return character['生命值'] > 0

player = {'名字': '林风',   '生命值': 45}
enemy  = {'名字': '暗影狼', '生命值': 0}

print(player['名字'], '还活着吗？', is_alive(player))
print(enemy['名字'],  '还活着吗？', is_alive(enemy))
```

#### 配合 if 判断使用

**这就非常实用了**

```python
def is_alive(character):
    return character['生命值'] > 0

player = {'名字': '林风', '生命值': 45}

if is_alive(player):
    print(player['名字'], '继续战斗！')
else:
    print(player['名字'], '已经倒下了……')
```
