## 第12课：字典让角色属性更清楚

#### 本节课的目的

前面我们已经学过变量、列表，也能做简单的背包和主菜单了。

但如果角色属性越来越多，光靠一个个单独变量来保存，就会慢慢变乱。

比如一个角色可能同时有：名字、职业、生命值、金币、武器、等级。

这一节课，我们要认识一种很适合保存“属性”的结构：字典 `dict`。

**这节课结束时，你应该能做到**

- 知道字典是用来保存“名称 -> 内容”对应关系的
- 会创建简单字典
- 会读取和修改字典里的值
- 能做一个角色属性系统 v1

#### 先理解字典是什么

**一句话理解**

字典就像一张信息表。

表里不是只按顺序摆东西，而是每个内容前面都有一个名字。

```python
player = {
    "名字": "林风",
    "职业": "剑士",
    "生命值": 100,
    "金币": 20
}

print(player)
```

#### 看懂这个字典

**先看外形**

- 字典用花括号 `{}`
- 左边是“名字”，右边是对应的内容
- 中间用冒号 `:` 连起来
- 每一组之间用逗号隔开

#### 怎样取出字典里的内容

**这一步很重要**

如果想查看角色的金币或生命值，就要通过对应的名字去取。

```python
player = {
    "名字": "林风",
    "职业": "剑士",
    "生命值": 100,
    "金币": 20
}

print("名字：", player["名字"])
print("生命值：", player["生命值"])
print("金币：", player["金币"])
```

#### 字典里的值也可以改变

**这就很适合拿来做属性变化**

```python
player = {
    "名字": "林风",
    "生命值": 100,
    "金币": 20
}

player["生命值"] = player["生命值"] - 15
player["金币"] = player["金币"] + 10

print(player)
```

#### 还可以增加新的属性

**角色会越来越完整**

```python
player = {
    "名字": "林风",
    "职业": "剑士"
}

player["武器"] = "铁剑"
player["等级"] = 1

print(player)
```
