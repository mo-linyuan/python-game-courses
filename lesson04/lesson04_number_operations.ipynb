#### 第一段示范代码

**先运行，再观察**

看看数字变量和平时的文字变量有什么不同。

```python
hp = 100
gold = 25
level = 1

print("生命值：", hp)
print("金币：", gold)
print("等级：", level)
```

#### 四个常用运算符

**今天先认识这四个符号**

- `+` 加法
- `-` 减法
- `*` 乘法
- `/` 除法

```python
print(3 + 2)
print(10 - 4)
print(5 * 3)
print(12 / 3)
```

#### 把运算放进游戏里

**这时候就能和游戏联系起来了**

比如角色受到攻击后，生命值会下降。

```python
hp = 100
damage = 25

hp = hp - damage

print("角色受到了攻击。")
print("现在的生命值：", hp)
```

#### 再看一个奖励的例子

**金币增加也是同样的道理**

```python
gold = 20
reward = 15

gold = gold + reward

print("你打开了宝箱。")
print("现在的金币：", gold)
```

#### 你来改一改

**任务**

```python
hp = 120
damage = 30
gold = 10
reward = 40

hp = hp - damage
gold = gold + reward

print("战斗结束后，生命值：", hp)
print("战斗结束后，金币：", gold)
```

#### 一个重要提醒

**这里可以顺手再强调一次**

`hp = hp - 30` 不是数学题里的“hp 等于 hp 减 30”。

它的意思是：先算出 `hp - 30` 的结果，再把这个新结果重新存回 `hp`。
