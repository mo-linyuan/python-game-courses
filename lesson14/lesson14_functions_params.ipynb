#### 先看问题出在哪里

**这个函数有什么缺点？**

````python
def show_status():
    print('名字：林风')
    print('生命值：100')
    print('金币：20')

show_status()
````

名字和数字都是写死的，只能显示林风，没办法显示其他角色。

所以我们需要让函数能接收外面传进来的值。

#### 什么是参数

**一句话理解**

参数就是函数括号里的「入口」。

你调用函数的时候，从外面把值传进去，函数内部就能用那个值。

````python
def greet(name):
    print('欢迎你，' + name + '！')

greet('林风')
greet('雷鸣')
greet('影月')
````

#### 看懂这段代码

**拆开来看**

- `def greet(name):` 里的 `name` 就是参数——它是一个占位符
- 调用 `greet('林风')` 时，`'林风'` 就被传给了 `name`
- 函数内部用 `name` 就能拿到那个值

同一个函数，传不同的值，就会有不同的结果。

#### 参数也可以是数字

**不只是文字，什么都能传**

````python
def show_hp(hp):
    print('当前生命值：', hp)

show_hp(100)
show_hp(45)
show_hp(0)
````

#### 函数可以有多个参数

**多个参数之间用逗号隔开**

````python
def show_status(name, hp, gold):
    print('====================')
    print('角色状态')
    print('名字：', name)
    print('生命值：', hp)
    print('金币：', gold)
    print('====================')

show_status('林风', 100, 20)
show_status('雷鸣', 80, 50)
````

#### 进一步：把字典传给函数

**回忆一下第12课**

我们已经学过用字典保存角色属性。

字典也可以作为参数传给函数！

````python
def show_player(player):
    print('====================')
    print('角色状态')
    print('名字：', player['名字'])
    print('职业：', player['职业'])
    print('生命值：', player['生命值'])
    print('金币：', player['金币'])
    print('====================')

player1 = {'名字': '林风', '职业': '剑士', '生命值': 100, '金币': 20}
player2 = {'名字': '雷鸣', '职业': '法师', '生命值': 80,  '金币': 50}

show_player(player1)
show_player(player2)
````
