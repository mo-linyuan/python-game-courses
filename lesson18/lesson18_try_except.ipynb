## 第18课：程序不崩溃的秘诀——try / except

#### 本节课的目的

做游戏绕不开一个问题：玩家可能乱输入。

比如你问玩家「请输入攻击次数」，他输了个「哈哈」。

这时候程序会直接崩溃，报一个 `ValueError`，游戏就结束了。

这节课我们学怎么让程序「犯错不崩」——用 `try / except` 抓住错误，友好地处理它。

**这节课结束时，你应该能做到**

- 知道什么情况下程序会报错崩溃
- 会用 `try / except` 抓住错误
- 能写一个「输错了就重新来」的输入循环
- 让游戏的输入环节更稳健

#### 先看程序是怎么崩的

**运行下面这段代码，输入一个非数字，看看会发生什么**

````python
answer = input('请输入你的攻击次数：')
times  = int(answer)   # 如果 answer 不是数字，这里就会崩
print(f'你发动了 {times} 次攻击！')
````

输入 `abc` 之后，你会看到 `ValueError`。这就是程序崩溃的样子。

#### try / except 是怎么工作的

**一句话理解**

`try` 里放「可能出错的代码」，`except` 里放「出错了怎么办」。

````python
try:
    answer = input('请输入攻击次数：')
    times  = int(answer)
    print(f'你发动了 {times} 次攻击！')
except ValueError:
    print('输入有误，请输入一个数字！')
````

#### 看懂这段代码

- `try:` 下面的代码会被「监控」着执行
- 如果出现了 `ValueError`，就跳到 `except ValueError:` 里
- 程序不会崩溃，而是继续往下走

#### 结合 while 做「不输对不让过」的循环

**这是游戏输入最常用的模式**

````python
while True:
    try:
        hp = int(input('请输入你的生命值（1-100）：'))
        if hp < 1 or hp > 100:
            print('范围不对，请输入 1 到 100 之间的数字。')
        else:
            print(f'好的，你的生命值设为 {hp}。')
            break
    except ValueError:
        print('那不是数字，再来一次。')
````

#### 用函数封装输入验证

**结合第13-15课学的函数，让它更整洁**

````python
def ask_number(prompt, min_val, max_val):
    while True:
        try:
            value = int(input(prompt))
            if min_val <= value <= max_val:
                return value
            else:
                print(f'请输入 {min_val} 到 {max_val} 之间的数字。')
        except ValueError:
            print('输入有误，请输入数字。')

hp     = ask_number('请设置生命值（1-100）：', 1, 100)
attack = ask_number('请设置攻击力（5-30）：',  5, 30)

print(f'角色已创建！HP：{hp}  攻击力：{attack}')
````

把输入验证做成函数后，以后每次需要让玩家输数字，直接调用 `ask_number()` 就行了。
