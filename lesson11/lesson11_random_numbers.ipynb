## 第11课：随机数让游戏变得不可预测

#### 本节课的目的

前面的游戏里，很多结果其实都是固定的。

如果玩家每次打开宝箱、每次前进、每次遇敌，结果都完全一样，游戏很快就会失去乐趣。

这一节课，我们要认识一个特别重要的工具：随机数。

有了随机数，游戏里就可以出现运气事件、掉宝系统、随机伤害、猜数字挑战。

**这节课结束时，你应该能做到**

- 知道什么是随机数
- 会使用 `random.randint()`
- 能做出随机奖励和随机伤害
- 能完成一个简单的猜数字小游戏

#### 第一步：导入随机工具

**今天会多一个新动作**

在 Python 里，随机数功能放在 `random` 这个工具包里。

要用它，先写：`import random`。

```python
import random

print(random.randint(1, 6))
```

#### 看懂这行代码

**先抓住最核心的地方**

- `random.randint(1, 6)` 的意思是：随机得到 1 到 6 之间的一个整数
- 每次运行，结果可能都不一样
- 这就像掷骰子一样

#### 多运行几次看看

**这一步一定要让学生自己试**

反复运行下面这格，观察输出是不是每次都一样。

```python
import random

print("你掷出了：", random.randint(1, 6))
```

#### 把随机数放进游戏里

**最常见的用法之一，就是随机奖励**

```python
import random

gold = random.randint(10, 50)
print("你打开宝箱，获得了", gold, "枚金币。")
```

#### 再看一个随机伤害的例子

**这时候就更像战斗了**

```python
import random

hp = 100
damage = random.randint(5, 20)
hp = hp - damage

print("怪物造成了", damage, "点伤害。")
print("你剩下的生命值：", hp)
```

#### 参考版本
