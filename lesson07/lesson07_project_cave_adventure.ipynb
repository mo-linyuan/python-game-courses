#### 第一步：先把游戏开场搭出来

**游戏从开场开始**

```python
print("====================")
print("《神秘洞穴大冒险》")
print("欢迎来到神秘洞穴。")
print("====================")
```

#### 第二步：让玩家进入游戏

**先拿到角色信息**

```python
name = input("请输入角色名字：")
print("欢迎你，" + name + "。")
```

#### 第三步：给角色一些初始属性

**游戏里的主角不能只有名字，还要有状态**

```python
hp = 100
gold = 20

print(name, "的生命值：", hp)
print(name, "的金币：", gold)
```

#### 第四步：加入洞穴里的选择

**这一步开始，游戏就真正动起来了**

```python
print("你走进洞穴深处，看见了三条路。")
choice = input("请选择左边、前方或右边：")
```

#### 第五步：根据不同路线改变结果

**这一步会用到上节课的多分支判断**

```python
if choice == "左边":
    gold = gold + 20
    print("你发现了一个宝箱，金币增加了。")
elif choice == "前方":
    hp = hp - 30
    print("你遇到了怪物，生命值减少了。")
else:
    hp = hp + 10
    print("你找到了一瓶药水，生命值恢复了。")
```

#### 第六步：把结果显示出来

**项目最后一定要有结算画面**

```python
print("====================")
print(name, "完成了这次探索。")
print("当前生命值：", hp)
print("当前金币：", gold)
print("====================")
```

#### 参考版本

**建议学生先做，再看这里**

```python
print("====================")
print("《神秘洞穴大冒险》")
print("欢迎来到神秘洞穴。")
print("====================")

name = input("请输入角色名字：")
hp = 100
gold = 20

print(name, "拿着火把走进了洞穴。")
print("你面前出现了三条路：左边、前方、右边。")
choice = input("请选择一条路：")

if choice == "左边":
    gold = gold + 20
    print("你找到了一只旧宝箱，金币增加了。")
elif choice == "前方":
    hp = hp - 30
    print("你遇到了一只怪物，生命值减少了。")
else:
    hp = hp + 10
    print("你发现了一瓶药水，生命值恢复了。")

print("====================")
print(name, "结束了这次探索。")
print("生命值：", hp)
print("金币：", gold)
print("====================")
```
