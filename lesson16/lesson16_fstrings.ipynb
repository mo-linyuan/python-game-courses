## 第16课：让文字更好看——f-string

#### 本节课的目的

到目前为止，我们打印信息都是这样做的：

```python
print('名字：', player['名字'], '生命值：', player['生命值'])
```

这种写法有点麻烦——逗号很多，空格也不好控制，拼出来的句子总是怪怪的。

这节课我们学一个更好用的写法：**f-string**。

以后写游戏台词和战斗播报，代码会变得清爽很多。

**这节课结束时，你应该能做到**

- 知道 f-string 是什么
- 会把变量嵌进字符串里
- 能改写之前用逗号拼接的 print 语句
- 能写出流畅的游戏台词和战斗播报

#### 先看旧写法的问题

**下面两段代码的输出结果一样，但哪个更好写？**

````python
name = '林风'
hp = 85

# 旧写法
print('勇士', name, '当前生命值为', hp, '点。')
````

````python
name = '林风'
hp = 85

# f-string 写法
print(f'勇士{name}当前生命值为{hp}点。')
````

两段代码的输出完全一样，但 f-string 更直观——就像在填空一样。

#### f-string 的两个规则

记住这两点就够了：

1. 字符串前面加一个 `f`
2. 想嵌入变量的地方用 `{}` 包起来

````python
hero  = '雷鸣'
level = 5
gold  = 120

print(f'{hero} 是 {level} 级勇士，身上有 {gold} 枚金币。')
````

#### 大括号里不只能放变量

**可以直接放计算式**

````python
attack  = 20
defense = 7

print(f'本次造成伤害：{attack - defense} 点')
print(f'双倍伤害是：{(attack - defense) * 2} 点')
````

#### 配合字典使用

**这是最常用的场景**

````python
player = {
    '名字': '林风',
    '职业': '剑士',
    '生命值': 85,
    '金币': 60
}

print(f"【{player['名字']}】职业：{player['职业']}  HP：{player['生命值']}  金币：{player['金币']}")
````

注意：字典在 f-string 的大括号里用**单引号**，因为外面的 f-string 已经用了双引号。
