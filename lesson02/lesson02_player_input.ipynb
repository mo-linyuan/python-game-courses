#### 本节课的目的

上一节课里，程序只能把我们提前写好的内容显示出来。

这一节课，我们要让程序开始“听”玩家说话，也就是学会接收输入。这样，程序就不再像一张海报，而更像一个会互动的小作品。

让程序接收输入。

#### 第一段互动代码

**先运行，再体验**

运行后，程序会先停下来等你输入。

```python
name = input("请输入你的名字：")
print("欢迎你，" + name + "！")
```

#### 看懂这段代码

**我们分 3 步看**

- `input()` 会弹出一个提问
- 你输入的内容会先被记住
- `print()` 再把它放进欢迎语里显示出来

**一句总结**

`input()` 是让程序接收用户输入。

#### 输入更多信息

**目标**

```python
name = input("请输入角色名字：")
age = input("请输入你的年龄：")
weapon = input("请选择你的武器：")

print("角色创建成功！")
print("名字：" + name)
print("年龄：" + age)
print("武器：" + weapon)
```

#### 输入、处理、输出

**这是本课最重要的编程思维**

- 输入：从用户那里拿到信息
- 处理：把信息整理、组合、安排进句子里
- 输出：把结果显示出来

**举个例子**

比如奶茶店点单也是这样：顾客说口味，这是输入；店员记下来并制作，这是处理；最后把奶茶递给你，这是输出。
