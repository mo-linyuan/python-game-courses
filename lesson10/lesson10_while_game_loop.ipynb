#### 先理解 `for` 和 `while` 的区别

**可以这样简单理解**

- `for`：已经知道要重复几次
- `while`：只知道“满足条件就继续”

所以 `while` 特别适合做“游戏还在继续”的场景。

#### 第一段示范代码

**先运行，再观察**

看看下面这段代码为什么会重复执行。

```python
energy = 3

while energy > 0:
    print("你又完成了一次训练。")
    energy = energy - 1

print("训练结束。")
```

#### 看懂这段代码

**这段代码要抓住两件事**

- `while energy > 0` 表示：只要体力还大于 0，就继续
- `energy = energy - 1` 很重要，因为它会让条件慢慢发生变化

如果条件永远不变，循环就可能一直跑下去。

#### 再看一个更像游戏的例子

**这次让玩家决定什么时候结束**

```python
command = input("输入 继续 开始冒险，输入 退出 结束游戏：")

while command != "退出":
    print("你继续向前探索。")
    command = input("输入 继续 或 退出：")

print("游戏结束。")
```

#### 这里为什么不会卡住

**关键就在这一行**

`command = input(...)`

因为每次循环都会重新问一次，所以玩家有机会输入“退出”，循环就会停下来。

#### `while` 和 `if` 一起用会更强

**这就是做游戏主菜单的基础**

我们可以让游戏一直运行，然后每一轮根据玩家输入做不同事。

```python
hp = 100
gold = 20
choice = ""

while choice != "退出":
    print("请选择：查看状态、休息、退出")
    choice = input("请输入操作：")

    if choice == "查看状态":
        print("生命值：", hp)
        print("金币：", gold)
    elif choice == "休息":
        hp = hp + 10
        print("你休息了一会儿，生命值恢复了。")

print("下次再来冒险吧。")
```
