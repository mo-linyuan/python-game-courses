## 第13课：函数让代码开始像工具箱

#### 本节课的目的

前面我们写了不少小游戏和系统，现在会慢慢遇到一个问题：有些代码会反复出现。

比如：

- 每次都要打印角色状态
- 每次战斗结束都要输出结算结果
- 每次进入关卡都要显示开场旁白

如果每次都复制一遍，代码会越来越长，也越来越乱。

这一节课，我们要认识一个很重要的工具：函数。

**这节课结束时，你应该能做到**

- 知道函数可以把重复代码打包起来
- 会定义最基础的函数
- 会调用函数
- 能把“显示角色状态”做成一个函数

#### 先看没有函数时会发生什么

**先感受一下为什么需要函数**

```python
print("====================")
print("角色状态")
print("名字：林风")
print("生命值：100")
print("金币：20")
print("====================")

print("====================")
print("角色状态")
print("名字：林风")
print("生命值：100")
print("金币：20")
print("====================")
```

这段代码虽然能运行，但明显有很多重复。

如果以后要改格式，就得改很多地方。

#### 什么是函数

**一句话理解**

函数就像一个可以反复使用的小工具。

你先把某段代码打包起来，给它起一个名字，以后想用的时候只要叫它一下就行。

```python
def say_hello():
    print("欢迎来到冒险世界。")

say_hello()
```

#### 看懂这段代码

**先抓住这两个动作**

- `def say_hello():` 表示“定义一个叫 `say_hello` 的函数”
- `say_hello()` 表示“调用这个函数”

定义是先把工具做好，调用才是真正使用它。

#### 函数可以反复调用

**这就是它的价值所在**

```python
def say_hello():
    print("欢迎来到冒险世界。")

say_hello()
say_hello()
say_hello()
```

#### 再看一个更有用的例子

**把角色状态打包成函数**

```python
def show_status():
    print("====================")
    print("角色状态")
    print("名字：林风")
    print("生命值：100")
    print("金币：20")
    print("====================")

show_status()
```
