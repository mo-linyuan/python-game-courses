#### 先理解什么叫判断

**可以先用生活例子讲**

- 如果下雨，就带伞
- 如果作业写完了，就可以休息
- 如果门开着，就走进去

编程里的 `if`，就是在做这种“如果……就……”的判断。

#### 第一段示范代码

**先运行，再观察**

看看当生命值变成 0 的时候，会发生什么。

```python
hp = 0

if hp <= 0:
    print("游戏结束。")
```

#### 看懂这段代码

**这一课最关键的格式要讲清楚**

- `if` 后面跟一个条件
- 这一行最后要有冒号 `:`
- 下一行前面要空 4 格，表示“如果条件成立，就执行这里”

#### 常见比较符号

**今天先认识几个常用的**

- `==` 表示“是否等于”
- `>=` 表示“大于等于”
- `<=` 表示“小于等于”
- `>` 表示“大于”
- `<` 表示“小于”

**特别提醒**

`=` 是赋值，`==` 才是比较。

```python
gold = 35

if gold >= 30:
    print("你有足够的金币，可以买药水。")
```

#### 再看一个和玩家选择有关的例子

**这时候就越来越像游戏了**

```python
choice = input("你要走左边还是右边？")

if choice == "左边":
    print("你走进了安静的森林小路。")
```

#### 如果不满足条件怎么办

**这时候就要用 `else`**

`else` 可以理解成“否则”。

```python
choice = input("你要走左边还是右边？")

if choice == "左边":
    print("你走进了安静的森林小路。")
else:
    print("你来到了一条通向山洞的小路。")
```
