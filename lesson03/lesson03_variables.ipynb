## 第3课：变量就是程序的小仓库

#### 本节课的目的

前两节课里，程序已经会显示内容，也会接收输入。

这一节课，我们要解决一个很关键的问题：程序怎样把信息记住，并且在后面继续使用？

这就要认识“变量”。

**这节课结束时，你应该能做到**

- 知道变量是干什么的
- 会用变量保存名字、职业、金币等信息
- 知道变量的值可以变化
- 能做出一个带角色信息的简单面板

#### 先用一个比喻理解变量

**可以这样理解**

变量就像一个贴了标签的小盒子。盒子外面写着名字，盒子里面装着内容。

比如叫 `name` 的盒子里装角色名字，叫 `gold` 的盒子里装金币数量。

#### 第一段示范代码

**先运行，再观察**

看看这些变量分别装了什么。

```python
name = "小勇者"
hp = 100
gold = 20

print(name)
print(hp)
print(gold)
```

#### 看懂变量长什么样

**可以拆开讲这句**

`name = "小勇者"`

- 左边 `name` 是变量名
- 右边 `"小勇者"` 是要保存的内容
- `=` 可以理解成“把右边的东西放进左边这个盒子里”

#### 用变量做一张角色卡

**这时候可以联系游戏经验**

很多游戏开场时都有角色信息面板，本质上就是把一堆变量显示出来。

```python
name = "林风"
job = "剑士"
hp = 120
gold = 30

print("角色信息")
print("名字：", name)
print("职业：", job)
print("生命值：", hp)
print("金币：", gold)
```

#### 变量的值可以改变

**这是本课另一个重点**

游戏里的数据不是永远不变的。受伤会掉血，打怪会加金币，升级会变强。

```python
hp = 100
gold = 10

hp = hp - 20
gold = gold + 15

print("现在的生命值：", hp)
print("现在的金币：", gold)
```

#### 这段变化可以怎么理解

**可以这样理解**

- 原来 `hp` 是 100
- `hp = hp - 20` 的意思不是数学课里的等号，而是“把新的结果重新存回 `hp`”
- 所以最后 `hp` 变成了 80

**生活类比**

像钱包里原本有 10 元，又收到了 15 元，现在就变成 25 元。

#### 把输入和变量连起来

**这一步很关键**

上一节课学了 `input()`，这一节课学了变量。现在把两者连起来，程序就能把玩家输入的信息保存下来。

```python
name = input("请输入角色名字：")
job = input("请输入职业：")
weapon = input("请输入武器：")

hp = 100
gold = 50

print("角色信息")
print("名字：", name)
print("职业：", job)
print("武器：", weapon)
print("生命值：", hp)
print("金币：", gold)
```
