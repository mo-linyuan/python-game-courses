## 第17课：列表装字典——管理多个角色和物品

#### 本节课的目的

到目前为止，我们的游戏只有一个玩家和一个敌人。

但真实的游戏里，可能会有一支队伍、一批敌人、一个装着很多物品的背包。

我们需要一种方法，能同时管理很多个字典。

这节课的答案是：**把字典放进列表**。

列表和字典我们都学过了，现在把它们组合起来用。

**这节课结束时，你应该能做到**

- 知道列表可以装字典
- 会用 for 循环遍历一批角色或物品
- 能做一个敌人波次系统和背包系统

#### 先回顾：列表 vs 字典

- 列表：一排东西，按顺序摆放
- 字典：一个东西的详细信息

把字典放进列表，就是「一排有详细信息的东西」。

#### 一个最简单的例子

**一支队伍**

```python
team = [
    {'名字': '林风', '职业': '剑士', '生命值': 100},
    {'名字': '雷鸣', '职业': '法师', '生命值': 80},
    {'名字': '影月', '职业': '刺客', '生命值': 90}
]

print(team)
```

#### 用索引取出其中一个角色

```python
team = [
    {'名字': '林风', '职业': '剑士', '生命值': 100},
    {'名字': '雷鸣', '职业': '法师', '生命值': 80},
    {'名字': '影月', '职业': '刺客', '生命值': 90}
]

first = team[0]
print(f"队伍第一个成员：{first['名字']}，{first['职业']}")
```

#### 用 for 循环遍历所有成员

**这才是最常用的操作**

```python
team = [
    {'名字': '林风', '职业': '剑士', '生命值': 100},
    {'名字': '雷鸣', '职业': '法师', '生命值': 80},
    {'名字': '影月', '职业': '刺客', '生命值': 90}
]

for member in team:
    print(f"{member['名字']}（{member['职业']}）- HP：{member['生命值']}")
```

#### 实际应用：敌人波次

**游戏里一波一波出现的敌人，就可以这样做**

```python
enemies = [
    {'名字': '哥布林',   '生命值': 30,  '攻击力': 8},
    {'名字': '暗影狼',   '生命值': 60,  '攻击力': 12},
    {'名字': '石头傀儡', '生命值': 120, '攻击力': 20}
]

print('=== 本关敌人 ===')
for i, enemy in enumerate(enemies):
    print(f"第{i+1}波：{enemy['名字']}  HP {enemy['生命值']}  ATK {enemy['攻击力']}")
```

`enumerate()` 会同时给你「序号」和「内容」，非常方便。

#### 实际应用：背包里的物品

**之前我们的背包只存了物品名字，现在可以存更多信息**

```python
backpack = [
    {'名称': '血瓶',   '类型': '消耗品', '效果': '回复 30 HP'},
    {'名称': '铁剑',   '类型': '武器',   '效果': '攻击力 +10'},
    {'名称': '皮甲',   '类型': '防具',   '效果': '防御力 +5'}
]

print('=== 背包 ===')
for item in backpack:
    print(f"  [{item['类型']}] {item['名称']} — {item['效果']}")
```

#### 在循环里修改数据

**让所有敌人的生命值减少**

```python
enemies = [
    {'名字': '哥布林A', '生命值': 30},
    {'名字': '哥布林B', '生命值': 30},
    {'名字': '哥布林C', '生命值': 30}
]

# 范围攻击，所有敌人各扣 10 点
for enemy in enemies:
    enemy['生命值'] -= 10

for enemy in enemies:
    print(f"{enemy['名字']} 剩余 HP：{enemy['生命值']}")
```

```python
enemies = [
    {'名字': '哥布林',   '生命值': 30,  '攻击力': 8},
    {'名字': '暗影狼',   '生命值': 60,  '攻击力': 12},
    {'名字': '石头傀儡', '生命值': 120, '攻击力': 20}
]

print('=== 出现的敌人 ===')
for e in enemies:
    print(f"  {e['名字']}  HP {e['生命值']}  ATK {e['攻击力']}")

print('\n=== 范围攻击，每只受 15 点伤害 ===')
for e in enemies:
    e['生命值'] -= 15

print('\n=== 攻击后状态 ===')
for e in enemies:
    status = '存活' if e['生命值'] > 0 else '已倒下'
    print(f"  {e['名字']}  剩余 HP：{e['生命值']}  [{status}]")
```
